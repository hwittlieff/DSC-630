## Assumptions

This analysis assumes that the L.A. Dodgers do not have control over their schedule (which day the games are played, weather conditions, etc.) and thus suggestions on changes these conditions will not be made. 

The team does have impact on wether or not promotions are given, and if fireworks are used. Since the team can impact these areas, suggestions will only be made on when to use them. 

## Load libraries and dataset

```python
# Load libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load dataset
df = pd.read_csv(r'Data/dodgers-2022.csv')

# View the dataframe
df.head(5)
```

## Clean dataset & add column for promotions

```python
# Convert cap to binary
df['cap'] = df['cap'].eq('YES').mul(1)
```

```python
# Convert shirt to binary
df['shirt'] = df['shirt'].eq('YES').mul(1)
```

```python
# Convert fireworks to binary
df['fireworks'] = df['fireworks'].eq('YES').mul(1)
```

```python
# Convert bobblehead to binary
df['bobblehead'] = df['bobblehead'].eq('YES').mul(1)
```

```python
# Create a new field for promotion
df['promotions'] = df['cap'] + df['shirt'] + df['bobblehead'] + df['fireworks']
```

```python
# Validate changes
df.head()
```

```python
# View the total number of promotions for any given game
df['promotions'].unique()
```

## Investigate promotions impact on attend

```python
# Extract data for promotions given
df_promotions=df.loc[df['promotions'] == 1]

# Extract data for no promotions
df_no_promotions=df.loc[df['promotions'] == 0]
```

```python
# Create weekday df for promotions
df_weekday_promo = df_promotions[['day_of_week', 'attend']]

# Groupby day of week and view summary statistics
print('Attendance for Promotions Given by Weekday')
df_weekday_promo.groupby('day_of_week').describe()
```

```python
# Create weekday df for no promotions
df_weekday_no_promo = df_no_promotions[['day_of_week', 'attend']]

# Groupby day of week and view summary statistics
print('Attendance for No Promotions Given by Weekday')
df_weekday_no_promo.groupby('day_of_week').describe()
```

```python
# Create dataframe for weekday no promo attendance means 
weekday_df1 = df_weekday_no_promo.groupby('day_of_week').mean()

# Rename Header
weekday_df1.columns = ['No Promotion Mean Attendance']

# View dataframe
weekday_df1
```

```python
# Create dataframe for weekday promo attendance means 
weekday_df2 = df_weekday_promo.groupby('day_of_week').mean()

# Rename Header
weekday_df2.columns = ['Promotion Mean Attendance']

# View dataframe
weekday_df2
```

```python
# Merge dataframes
weekday_df = weekday_df2.merge(weekday_df1, left_index=True, right_index=True)

# Add column for increase in mean attendance
weekday_df['Mean Attendance Increase'] = weekday_df['Promotion Mean Attendance'] - weekday_df['No Promotion Mean Attendance']

# Reset index
weekday_df = weekday_df.reset_index()

# Create a sorted df by increase in attendance
sorted_df = weekday_df.sort_values(by=['Mean Attendance Increase'], ascending=False)

# Display the df
sorted_df
```

```python
# Build a barchart to display the changes in attendance when using promotions by weekday
sorted_df.plot(x="day_of_week", y=['Promotion Mean Attendance', 'No Promotion Mean Attendance'], kind='bar')
```

## Summary

Promotions have a great impact on the overall attendance for Dodger's games. In order to obtain the most value from promotions, promotions should be utilized primarily for games on Thursdays. Promotions can then be expanded to Wednesdays and Mondays based on funding availability.

## 1. Write a summary of your data and identify at least two questions to explore visually with your data.

The "tesla-dataset" from Kaggle contains details for the past 5 years on Tesla's Stock. These details include metrics on price and volume by day for Tesla's stock. The dataset can be found via the link below.

https://www.kaggle.com/surajjha101/tesla-share-price-for-last-5-years

With this dataset several questions can be investigated about Tesla's stock including:
<ol>
    <li>Does volume traded correlate with the change in stock price?</li>
    <li>What is the typical volume traded for a day?</li>
</ol>

```python
# Load libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load dataset
df = pd.read_csv(r'Data/Tesla-dataset.csv')

# Transform date
df['Date'] = pd.to_datetime(df['Date'])

# Create the change in price
df['Gain/Loss'] = df['Close'] - df['Open']

# View the first five rows
df.head(5)
```

## 2. Create a histogram or bar graph from your data

```python
# Create a bar Chart of the daily volume
plt.bar(df['Date'], df['Volume'])
plt.xlabel('Date')
plt.ylabel('Volume Traded')
plt.title('Tesla Stock Volume by Date')
plt.show()
```

## 3. Create a boxplot from your data

```python
# Create a boxplot of volume
plt.boxplot(df['Volume'])
plt.title('Tesla Stock Trading Volume')
plt.show()
```

## 4. Create a bivariate plot from your data

```python
# Create a scatter plot of volume vs close
fig=plt.figure()
ax=fig.add_axes([0, 0, 1, 1])
ax.scatter(df['Volume'], df['Close'], color='r')
ax.set_xlabel('Volume Traded')
ax.set_ylabel('Closing Price')
ax.set_title('Tesla Stock Trading Volume vs Closing Price')
plt.show()
```

## 5. Create any additional visualizations that will help to answer the question(s) you want to answer.

```python
# Create a scatter plot of volume vs open
fig=plt.figure()
ax=fig.add_axes([0, 0, 1, 1])
ax.scatter(df['Volume'], df['Open'], color='b')
ax.set_xlabel('Volume Traded')
ax.set_ylabel('Opening Price')
ax.set_title('Tesla Stock Trading Volume vs Opening Price')
plt.show()
```

```python
# Create a scatter plot of gain/loss vs volume
fig=plt.figure()
ax=fig.add_axes([0, 0, 1, 1])
ax.scatter(df['Volume'], df['Gain/Loss'], color='g')
ax.set_xlabel('Volume Traded')
ax.set_ylabel('Gain/Loss')
ax.set_title('Tesla Stock Trading Volume vs Price Change')
plt.show()
```
